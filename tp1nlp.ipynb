 ---Mise en minuscules

```python
import pandas as pd

# Charger le fichier twcs.csv dans un DataFrame
df = pd.read_csv("twcs/twcs.csv")

# Sélectionner les colonnes 'tweet_id' et 'text'
df = df[['tweet_id', 'text']]

# Fonction pour mettre le texte en minuscules et nettoyer
def clean_text(text):
    return text.lower()

# Appliquer la fonction pour mettre le texte en minuscules et nettoyer à la colonne 'text'
df['text_cleaned'] = df['text'].apply(clean_text)

# Afficher les colonnes 'tweet_id', 'text' et 'text_cleaned' côte à côte
print(df[['tweet_id', 'text', 'text_cleaned']].head())
```

• Suppression des ponctuations

```python
import pandas as pd
import re

# Charger le fichier twcs.csv dans un DataFrame
df = pd.read_csv("twcs/twcs.csv")

# Fonction pour supprimer les ponctuations
def remove_punctuation(text):
    # Utiliser une expression régulière pour remplacer les ponctuations par une chaîne vide
    return re.sub(r'[^\w\s]', '', text)

# Appliquer la fonction de suppression des ponctuations à la colonne "text"
df['text_cleaned'] = df['text'].apply(remove_punctuation)

# Afficher les premières lignes du DataFrame avec les colonnes tweet_id, text et text_cleaned
print(df[['tweet_id', 'text', 'text_cleaned']].head())
```

*.---Suppression des mots vides (stopwords)*

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

# Charger le fichier twcs.csv dans un DataFrame
df = pd.read_csv("twcs/twcs.csv")

# Sélectionner uniquement la colonne "text"
texts = df['text']

# Liste des stopwords pour la langue anglaise
stop_words = set(stopwords.words('english'))

# Fonction pour supprimer les stopwords
def remove_stopwords(text):
    # Séparer le texte en mots
    words = text.split()
    # Retirer les stopwords
    filtered_words = [word for word in words if word.lower() not in stop_words]
    # Reconstruire le texte sans les stopwords
    return ' '.join(filtered_words)

# Appliquer la fonction remove_stopwords à chaque texte dans la colonne "text"
df['text_cleaned'] = texts.apply(remove_stopwords)

# Afficher les premières lignes du DataFrame avec les colonnes tweet_id, text et text_cleaned
print(df[['tweet_id', 'text', 'text_cleaned']].head())
```

---Suppression des mots fréquents : Identifiez et supprimez les mots extrêmement fréquents qui
pourraient être peu informatifs.

```python
import pandas as pd

# Charger le fichier twcs.csv dans un DataFrame
df = pd.read_csv("twcs/twcs.csv")

# Sélectionner uniquement la colonne "text"
texts = df['text']

# Calculer la fréquence des mots
word_freq = pd.Series(' '.join(texts).split()).value_counts()

# Définir un seuil pour les mots fréquents à supprimer (par exemple, les 10 mots les plus fréquents)
threshold = 10
most_common_words = word_freq[:threshold].index

# Fonction pour supprimer les mots fréquents
def remove_common_words(text):
    words = text.split()
    filtered_words = [word for word in words if word not in most_common_words]
    return ' '.join(filtered_words)

# Appliquer la fonction pour supprimer les mots fréquents à chaque texte dans la colonne "text"
df['text_cleaned'] = texts.apply(remove_common_words)

# Afficher les premières lignes du DataFrame avec les colonnes tweet_id, text et text_cleaned
print(df[['tweet_id', 'text', 'text_cleaned']].head())
```

Suppression des mots rares : Éliminez les mots qui apparaissent très rarement.
vous pouvez le faire :

Calculer la fréquence des mots dans la colonne "text".
Identifier les mots rares.
Supprimer ces mots de la colonne "text".

```python
import pandas as pd

# Charger le fichier twcs.csv dans un DataFrame
df = pd.read_csv("twcs/twcs.csv")

# Sélectionner uniquement la colonne "text"
texts = df['text']

# Calculer la fréquence des mots
word_freq = pd.Series(' '.join(texts).split()).value_counts()

# Définir un seuil pour les mots rares à supprimer (par exemple, les mots qui apparaissent moins de 1 fois)
threshold = 1
rare_words = word_freq[word_freq <= threshold].index

# Fonction pour supprimer les mots rares
def remove_rare_words(text):
    words = text.split()
    filtered_words = [word for word in words if word not in rare_words]
    return ' '.join(filtered_words)

# Appliquer la fonction pour supprimer les mots rares à chaque texte dans la colonne "text"
df['text_cleaned'] = texts.apply(remove_rare_words)

# Afficher les premières lignes du DataFrame avec les colonnes tweet_id, text et text_cleaned
print(df[['tweet_id', 'text', 'text_cleaned']].head())
```

Stemming : Réduisez les mots à leur racine pour simplifier le texte.

```python
import pandas as pd
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')

# Charger le fichier twcs.csv dans un DataFrame
df = pd.read_csv("twcs/twcs.csv")

# Sélectionner uniquement la colonne "text"
texts = df['text']

# Initialiser le stemmer Porter
stemmer = PorterStemmer()

# Fonction pour appliquer le stemming à chaque texte
def stem_text(text):
    # Tokeniser le texte en mots
    words = word_tokenize(text)
    # Appliquer le stemming à chaque mot et rejoindre les mots stemmés
    stemmed_words = [stemmer.stem(word) for word in words]
    return ' '.join(stemmed_words)

# Appliquer la fonction pour appliquer le stemming à chaque texte dans la colonne "text"
df['text_stemmed'] = texts.apply(stem_text)

# Afficher les premières lignes du DataFrame avec les colonnes tweet_id, text et text_stemmed
print(df[['tweet_id', 'text', 'text_stemmed']].head())
```

Lemmatisation : Convertissez les mots à leur forme de base en utilisant le contexte lexical.

```python
import pandas as pd
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
nltk.download('wordnet')

# Charger le fichier twcs.csv dans un DataFrame
df = pd.read_csv("twcs/twcs.csv")

# Sélectionner uniquement la colonne "text"
texts = df['text']

# Initialiser le lemmatizer WordNet
lemmatizer = WordNetLemmatizer()

# Fonction pour appliquer la lemmatisation à chaque texte
def lemmatize_text(text):
    # Tokeniser le texte en mots
    words = word_tokenize(text)
    # Appliquer la lemmatisation à chaque mot et rejoindre les mots lemmatisés
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    return ' '.join(lemmatized_words)

# Appliquer la fonction pour appliquer la lemmatisation à chaque texte dans la colonne "text"
df['text_lemmatized'] = texts.apply(lemmatize_text)

# Afficher les premières lignes du DataFrame avec les colonnes tweet_id, text et text_lemmatized
print(df[['tweet_id', 'text', 'text_lemmatized']].head())
```

Suppression des émojis , Suppression des émoticônes, Suppression des URL, Suppression des balises
HTML.

```python
import pandas as pd
import re

# Charger le fichier twcs.csv dans un DataFrame
df = pd.read_csv("twcs/twcs.csv")

# Sélectionner uniquement la colonne "text"
texts = df['text']

# Fonction pour supprimer les emojis, les émoticônes, les URL et les balises HTML
def remove_special_characters(text):
    # Supprimer les emojis
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    # Supprimer les émoticônes
    text = re.sub(r'[\U00010000-\U0010ffff]+', '', text)
    # Supprimer les URL
    text = re.sub(r'http\S+|www\S+', '', text)
    # Supprimer les balises HTML
    text = re.sub(r'<.*?>', '', text)
    return text

# Appliquer la fonction pour supprimer les emojis, les émoticônes, les URL et les balises HTML à chaque texte dans la colonne "text"
df['text_cleaned'] = texts.apply(remove_special_characters)

# Afficher les premières lignes du DataFrame avec les colonnes tweet_id, text et text_cleaned
print(df[['tweet_id', 'text', 'text_cleaned']].head())
```
